Question 1 : What is Dimensionality Reduction? Why is it important in machine
learning?
- Dimensionality reduction is the process of reducing the number of input variables (features) in a dataset while preserving as much relevant information as possible.
It transforms high-dimensional data into a lower-dimensional representation.
- Common techniques:
Feature Selection (e.g., removing irrelevant or redundant features)
Feature Extraction (e.g., PCA, t-SNE, LDA)
- Reduces Overfitting
- Improves Training Speed
- Enhances Model Performance
-  Helps Visualization
-  Eliminates Multicollinearity
- Suppose a dataset has 200 features, but only 20 contribute significantly to predictions.
Dimensionality reduction can help transform the data into those 20 meaningful features.


Question 2: Name and briefly describe three common dimensionality reduction
technique?

-1)Principal Component Analysis (PCA):
  - A linear dimensionality reduction method.
  - Transforms original features into new uncorrelated variables called principal components.
  - These components capture the maximum variance in the data.
  - Mostly used for feature extraction and visualization.

-2)t-SNE (t-Distributed Stochastic Neighbor Embedding):
  - A non-linear technique mainly used for visualizing high-dimensional data in 2D or 3D.
  - Preserves the local structure of the data, making it great for clustering.
  - Often used in image, NLP, and deep learning datasets.

-3)LDA (Linear Discriminant Analysis):
  - A supervised method (requires class labels).
  - Reduces dimensionality by maximizing class separability.
  - Often used in classification tasks like face recognition and sentiment analysis.

Question 3: What is clustering in unsupervised learning? Mention three popular
clustering algorithms.
- Clustering is an unsupervised learning technique used to group similar data points into clusters based on their characteristics — without using labeled data.
- The goal is to ensure:

High similarity within a cluster

Low similarity between different clusters

It helps in pattern discovery, grouping, and data exploration.
- Three Popular Clustering Algorithms:
  - 1)K-Means: Divides data into K clusters by minimizing within-cluster variance
  - 2)DBSCAN: Forms clusters based on dense regions and detects noise/outliers
  - 3)Hierarchical Clustering: Builds a tree (dendrogram) of clusters, no need to predefine K
  

Question 4: Explain the concept of anomaly detection and its significance.
- Anomaly detection (also called outlier detection) is the process of identifying rare or unusual data points that differ significantly from the majority of the data.
- These anomalies may indicate:

Errors/noise in data

Fraudulent activity

System failure

Novel or unexpected patterns

- Types of Anomalies: 1)Point anomalies: Single data point is very different from the rest.
2)Contextual anomalies: Anomalous based on context (e.g., temperature vs. time).
3)Collective anomalies: A group of points together form an unusual pattern.

- Significance of Anomaly Detection:
1)Fraud Detection: Detects unusual transactions in financial systems.
2)Cybersecurity:  Identifies network intrusions or malware
3)Fault Detection:  Helps in predictive maintenance of machines/sensors
4)Data Cleaning: Removes noisy/outlier data before model training
5)Better Decision-Making:  Highlights critical patterns that need attention

-Examples: Detecting credit card fraud using transaction data

Monitoring server health in IT systems

Identifying diseases using medical data

Spotting abnormal behavior in surveillance footage


Question 5: List and briefly describe three types of anomaly detection techniques.
- 1. Statistical Methods:
  - Based on probability and statistical models.
  - Assumes normal behavior follows a specific distribution (e.g., Gaussian).- Anomalies are detected when data points deviate significantly from the expected statistical range.
  - Example: Z-score, Gaussian Mixture Models (GMM).

- 2. Machine Learning-Based Methods:
  - Use algorithms to learn patterns in data and identify deviations.
  - Can be supervised (with labeled anomalies) or unsupervised (no labels).
  - Example: Isolation Forest, One-Class SVM, K-Means clustering for anomaly detection.

- 3. Density-Based Methods:
  - Detect anomalies by examining the density of data points.
  - Points in low-density regions (far from clusters) are considered anomalies.
  - Example: DBSCAN (Density-Based Spatial Clustering), LOF (Local Outlier Factor).

Question 6: What is time series analysis? Mention two key components of time series data?
- Time series analysis is the process of analyzing data points collected over time to identify patterns, trends, and relationships. It is widely used to forecast future values, understand historical behavior, and detect anomalies in sequential data such as stock prices, weather data, sales figures, sensor data, etc.
- Two Key Components of Time Series Data:

1)Trend: The overall long-term direction of the data (upward, downward, or stable).
Example: Increasing sales over several years.

2)Seasonality: Regular and repeating patterns that occur within fixed periods.
Example: Higher ice-cream sales every summer.

Question 7: Describe the difference between seasonality and cyclic behavior in time series?
- Seasonality: 1)Repeating patterns that occur at regular and fixed intervals. 2)Time Period: Always fixed (e.g., daily, monthly, yearly). 3)Often driven by external and predictable factors (e.g., weather, festivals, business cycles). 4)example: Retail sales increase every December (holiday season). 5)Seasonality is regular and predictable.

- Cyclic Behavior: 1)Fluctuations that occur over long and irregular periods. 2)Time Period: Not fixed – duration varies. 3)Caused by economic or natural factors that are harder to predict. 4)example: Economic recession or business expansion phases over several years. 5)Cyclic behavior is irregular and uncertain.

```python
"""
Question 8: Write Python code to perform K-means clustering on a sample dataset.
"""

import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


data = {
    'X': [1, 2, 3, 8, 9, 10],
    'Y': [1, 2, 3, 8, 9, 10]
}
df = pd.DataFrame(data)

kmeans = KMeans(n_clusters=2, random_state=0)  # Choose number of clusters (e.g., 2)
kmeans.fit(df)


df['Cluster'] = kmeans.labels_

plt.scatter(df['X'], df['Y'], c=df['Cluster'], s=100)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            marker='X', s=200, label='Centroids')
plt.title('K-means Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# final dataframe
print(df)
```

```python
"""
Question 9: What is inheritance in OOP? Provide a simple example in Python.
- Inheritance is a core concept in OOP where a child class (subclass) can inherit properties and behaviors (methods) from a parent class (superclass).
It promotes code reusability, modularity, and extensibility.

"""


class Animal:
    def sound(self):
        return "Animals make sounds"

class Dog(Animal):
    def sound(self):
        return "Dog barks"


a = Animal()
d = Dog()

print(a.sound())  # Output: Animals make sounds
print(d.sound())  # Output: Dog barks
```

```python
"""
Question 10: How can time series analysis be used for anomaly detection?
- Time series analysis is widely used for detecting unusual or unexpected patterns in data over time. These anomalies may indicate system failures, fraud, sensor errors, or rare events.
- Time series analysis helps by:

1)Modeling normal behavior over time (trend, seasonality, patterns).

2)Detecting deviations from expected values using statistical or machine learning techniques.

3)Flagging data points that significantly differ from predictions.
"""

#example:

import pandas as pd
import numpy as np


data = pd.DataFrame({
    'value': [10, 12, 11, 13, 100, 14, 12, 11, 13]  # 100 is an anomaly
})

data['rolling_mean'] = data['value'].rolling(window=3).mean()
data['rolling_std'] = data['value'].rolling(window=3).std()

# Detect anomaly
threshold = 2
data['anomaly'] = np.where(
    abs(data['value'] - data['rolling_mean']) > threshold * data['rolling_std'], "Yes", "No"
)

print(data)
```
